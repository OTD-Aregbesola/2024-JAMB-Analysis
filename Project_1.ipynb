## Bivariate Analysis


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting only numeric columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Plotting the heatmap
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

From the data above we can see that;

1. There's a strong correlation between JAMB Scores and Assignments completed.
2. There's a weak correlation between JAMB Scores and study hour per week
3. There's a weak correlation between JAMB Scores and Teacher Quality
4. There's a weak correlation between JAMB Scores and Distance to school
5. There's a weak correlation between JAMB Scores and Assignments completed

```python
# Visual visualization for JAMB score and Attendance rate

sns.lineplot( x = 'Attendance_Rate', y = 'JAMB_Score', data = df)

plt.title("Attendance Rate vs JAMB score")
plt.xlabel('Attendance_rate')
plt.ylabel('JAMB Score')
plt.show()
```
